```python
X = data.drop(['selling_price', 'owner'], axis=1)
y= data.selling_price
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(RandomForestRegressor(), {
    'n_estimators':[int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)],
    'max_features':['auto', 'sqrt'],
    'max_depth':[int(x) for x in np.linspace(5,30,num = 6)],
    'min_samples_split':[2,5,10,15,100],
    'min_samples_leaf':[1,2,5,10]})
```

```python
rs.fit(X_train, y_train)
```

```python
rs.best_params_
```

```python
param = pd.DataFrame(rs.cv_results_)
param
```

```python
param[['param_n_estimators', 'param_min_samples_split', 'param_min_samples_leaf', 'param_max_features','param_max_depth','mean_test_score']]
```

```python
model = RandomForestRegressor(n_estimators = 1000, min_samples_split = 2, min_samples_leaf = 2, max_features = 'auto', max_depth = 15)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn import metrics
pred = model.predict(X_test)
metrics.r2_score(pred, y_test)
```
